# Sentiment by genre and artist

```
import pandas as pd
import json
```

```
df_artist_info = pd.read_pickle("../../data/other_files/df_artist_info.pkl")
```

```
sentiment = []
```

```
for genre in df_artist_info.genre.unique():
    item = {
        "genre": genre,
        "happiest": [],
        "saddest": []
    }

    temp_df = df_artist_info[df_artist_info['genre'] == genre]

    saddest = temp_df.sort_values(by=['sentiment'],ascending=True)[0:5][['artist','sentiment']]
    saddest_key = saddest["artist"].tolist()
    saddest_value = saddest["sentiment"].tolist()

    happiest = temp_df.sort_values(by=['sentiment'],ascending=False)[0:5][['artist','sentiment']]
    happiest_key = happiest["artist"].tolist()
    happiest_value = happiest["sentiment"].tolist()

    for i in range(0, 5):
        item["saddest"].append({
            "name": saddest_key[i],
            "sentiment": saddest_value[i]
        })

        item["happiest"].append({
            "name": happiest_key[i],
            "sentiment": happiest_value[i]
        })

    sentiment.append(item)
```

```
sentiment
```

```
with open("../../data/webpage/sentiment-genre-artist.json", "w") as outfile:
    json.dump(sentiment, outfile)
```
