```
# Imports
import os
import json
import numpy as np
import matplotlib.pyplot as plt
```

```
with open("../../data/network.json") as f:
  artists = json.load(f)
```

```
popularity = []
for artist in artists:
    if artist["id"]:
        popularity.append(artist["popularity"])
```

```
labels = sorted(set(popularity))
counts = []

for l in labels:
    c = popularity.count(l)
    counts.append(c)
```

```
clean = [x for x in artists if x["id"]]
popularities = [x["popularity"] for x in clean if len(x["genres"]) > 0]
```

```
mean = np.mean(popularities)
mean
```

```

plt.bar(labels[:int(mean)], counts[:int(mean)], color="blue")
plt.bar(labels[int(mean):], counts[int(mean):], color="red")

plt.title("Artists' Popularity Distribution")
plt.xlabel('Popularity')
plt.ylabel('Number of artists')

plt.savefig("../../images/popularity-distribution.png", dpi=600, format="png")

plt.show()
```
